### Importación de librerías necesarias para  el proyecto

1. Se importan todas las librerías necesarias para realizar nuestro ETL

```python
import pandas as pd
from pandas import json_normalize
import numpy as np
import json as js
import ast as ast
```

2. Transformación de dataset "output_steam_games" a dataframe para manipulación con librería PANDAS

```python
games_def = pd.read_json('output_steam_games.json', lines=True) #lines=True le indica a la librería de Pandas que cada línea corresponde a un registro por separado
games_def
```

3. Eliminación de todas las filas donde todas sus columnas sean valores None

```python
games_def = games_def.dropna(how='all') #Con dropna se pueden eliminar todas las filas con una condición que es how='all', es decir, que solo eliminará registros cuando toda la fila tenga vacías sus columnas
games_def
```

4. Se eliminan las columnas que no tienen información relevante para las funciones que aplicaremos en el proyecto

```python
games_def = games_def.drop(columns=['publisher', 'title', 'url', 'reviews_url'])
games_def
```

5. Se comienza a revisar cuál es el tipo de dato con respecto a cada columna del dataframe

```python
games_def.dtypes    #dtypes permite ver el tipo de dato de cada columna dentro de mi dataframe
```

6. Se procederá a transformar la columna 'price' a un tipo de dato numérico, para ello se pasa a reemplazar los valores únicos que encontremos con la función .unique

```python
valores_unicos = games_def['price'].unique()    #unique permite mostrar  los valores únicos de una columna.
valores_unicos
```

7. Una vez identificados los elementos que constituyen la columna 'price', se procede a cambiarlos de acuerdo
al contexto en que se encuentren, en este caso, todos los elementos encontrados hacen referencia a que el 
precio del vídeo juego o producto del catalogo es gratis

```python
games_def['price'] = games_def['price'].replace('Free To Play', 0)
games_def['price'] = games_def['price'].replace('Free to Play', 0)
games_def['price'] = games_def['price'].replace('Free', 0)
games_def['price'] = games_def['price'].replace('Play for Free!', 0)
games_def['price'] = games_def['price'].replace('Free Mod', 0)
games_def['price'] = games_def['price'].replace('Install Now', 0)
games_def['price'] = games_def['price'].replace('Play the Demo', 0)
games_def['price'] = games_def['price'].replace('Starting at $499.00', 499)
games_def['price'] = games_def['price'].replace('Starting at $449.00', 449)
games_def['price'] = games_def['price'].replace('Free to Use', 0)
games_def['price'] = games_def['price'].replace('Free to try', 0)
games_def['price'] = games_def['price'].replace('Free Movie', 0)
games_def['price'] = games_def['price'].replace('Free HITMAN™ Holiday Pack', 0)
games_def['price'] = games_def['price'].replace('Free Demo', 0)
games_def['price'] = games_def['price'].replace('Third-party', 0)
games_def['price'] = games_def['price'].replace('Play WARMACHINE: Tactics Demo', 0)
games_def['price'] = games_def['price'].replace('Play Now', 0)
games_def['price'] = games_def['price'].replace('Free to Try', 0)
games_def['price'] = games_def['price'].replace('Install Theme', 0)

# Se selecciona el elemento contenido en la columna 'price', este argumento está contenido dentro de la función .replace, hacemos esto con todos los 
# elementos  que se encuentran en la columna 'price' '''
```

8. Se verifican que los valores de la columna 'price' haya quedado solo con valores númericos

```python
games_def.info()
```


8. Se convierten los elementos en la columna de fecha a 'Not a time', lo que permitirá identificar los problemas sin que se genere un error.

```python
games_def['release_date'] = pd.to_datetime(games_def['release_date'], errors='coerce') #El método .to_datetime convierte  la columna 'release_date' a formato datetime,  
                                                                                        #si encuentra algún valor que no pueda convertir a datetime, lo reemplaza con NaT
```

8. Se muestra un filtro con las filas donde no se pudieron acomodar los datos como formato de fecha

```python
games_def[games_def['release_date'].isna()] #El método isna, devolverá una serie 
                                            #booleana que indica  si el valor es NaN o no.
```

9. Se eliminan todos los registros donde aparecen estos datos donde la fecha no se puede acomodar al formato deseado

```python
games_def = games_def.dropna(subset=['release_date'])

# El método permitirá eliminar valores faltantes en la columna release_date 
```

10. Se verifica que la columna release date tiene  un tipo de datos de fecha y hora (datetime)

```python
games_def.info() 
```

11. Se crea una nueva columna que permita realizar una mejor manipulación de los datos, en este caso se crea una columna denominada "year"

```python
games_def['year'] = games_def['release_date'].dt.year 

# .dt. es una propiedad que permite acceder de una manera particular
#a  los datos de una columna de tipo datetime.
```

12. Aquí se reorganizan las columnas para que 'year' quede justo al lado de 'release_date'

```python
games_def = games_def.reindex(columns=['genres','app_name','release_date','year','tags','specs','price','early_access','id','developer'])
games_def
```

13. Se seleccionan y se suman todos  los valores nulos de cada columna del dataframe

```python
games_def.isnull().sum()

#El método isnull generará una respuesta booleana  para cada valor en la columna. 
# Sumar esto con .sum() nos da el número de valores nulos en cada  columna.
```

14. Al observar los datos nulos presentes en las columnas del dataframe, se determina que para efectos de la limpieza, se eliminarán las filas  con valores nulos en  todas las columnas

```python
games_def = games_def.dropna().reset_index(drop=True)
games_def
```

15. Renombraremos algunas columnas, para poder realizar un mejor relacionamiento con los otros datasets

```python
games_def.rename(columns={'id':'item_id','app_name':'item_name'},inplace=True)
```

16. Para efectos de un EDA más completo, vamos a desmontar la lista de strings de la columna 'genres', y escogeremos la primera categoria de vídeojuegos que aparece categorizada.

```python
games_def['genres'] = games_def['genres'].str[0]
```

17. Se identifica finalmente a través de una columna de valor únicos, como 'id'  o 'app_name si existen registros duplicados

```python
duplicados = games_def.duplicated(subset='item_id')
filas_duplicadas = games_def[duplicados]
filas_duplicadas
```

18. Se procede a eliminar el/los registro/s que estén duplicados tomando el id como columna de comparación.

```python
games_def.drop_duplicates(subset='item_id', inplace=True)
games_def.reset_index(drop=True, inplace=True) 

#El argumento drop=True se refiere a que los nuevos índices creados 
#van a reemplazar a los anteriores sobreescribiéndolos.
```

19. Una vez transformada  la información, se puede utilizar para realizar análisis de datos, crear informes y tomar decisiones, se verifica por última vez  que la información se encuentra en el lugar correcto y que se ha realizado la transformación de datos

```python
games_def.info()
```

20. Como última instancia, se procede a generar el archivo en formato .parquet, archivo que pesa poco y  que se puede leer rápidamente.

```python
games_def.to_parquet('ETL.GAMES.parquet', index=False)

# dentro del argumento el index = False,  
# se indica que no se debe incluir la columna de índice en el archivo parquet.
```
